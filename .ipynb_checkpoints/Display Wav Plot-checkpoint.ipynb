```python
import librosa
import matplotlib.pyplot as plt
import numpy as np

def waveplot(data, sr, emotion):
    time = np.arange(0, len(data)) / sr
    plt.figure(figsize=(10, 4))
    plt.plot(time, data, lw=1)
    plt.title(emotion, size=20)
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.show()

# Load audio data
audio_file_path = r"background_sound/13.wav"
data, sample_rate = librosa.load(audio_file_path)

# Display waveplot
# waveplot(data, sample_rate, "Vasudev")
```

```python
import os
import librosa
import matplotlib.pyplot as plt
import numpy as np

def waveplot_subplot(data, sr, emotion, ax):
    time = np.arange(0, len(data)) / sr
    ax.plot(time, data, lw=1)
    ax.set_title(emotion, size=10)
    ax.set_xlabel('Time (s)')
    ax.set_ylabel('Amplitude')

# Directory containing your audio files
directory_path = r"audio_data"

# Get all files in the directory
audio_files = [f for f in os.listdir(directory_path) if f.endswith('.wav')]

# Set up the subplots
num_rows = len(audio_files) // 6 + (len(audio_files) % 6 > 0)  # Calculate the number of rows
num_cols = min(len(audio_files), 6)  # Maximum of 6 columns

fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 2*num_rows))

# Iterate through the files and display waveplots
for i, audio_file in enumerate(audio_files):
    file_path = os.path.join(directory_path, audio_file)
    data, sample_rate = librosa.load(file_path)
    
    if num_rows == 1:  # If there is only one row, axs will not be a 2D array
        waveplot_subplot(data, sample_rate, audio_file, axs[i])
    else:
        row_idx, col_idx = divmod(i, 6)
        waveplot_subplot(data, sample_rate, audio_file, axs[row_idx, col_idx])

# Adjust layout to prevent clipping of titles
plt.tight_layout()

# Show the plot
plt.show()
```
